```python
import numpy as np
import pandas as pd
```

```python
true = pd.read_csv('True.csv', on_bad_lines='skip', engine='python')
fake = pd.read_csv('Fake.csv', on_bad_lines='skip', engine='python')
```

```python
true['label']=1
```

```python
fake['label']=0
```

```python
news= pd.concat([fake,true] , axis =0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title', 'subject', 'date'], axis=1, errors='ignore')
```

```python
news.head()
```

```python
news.tail()
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace =True)
```

```python
news.head()
```

```python
news.drop(['index'],axis =1, inplace =True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):
  text=text.lower()
  # Remove URLs
  text = re.sub(r'https?://\S+|www\.\S+', '', text)

# Remove HTML tags
  text = re.sub(r'<.*?>', '', text)

# Remove punctuation
  text = re.sub(r'[^\w\s]', '', text)

# Remove digits
  text = re.sub(r'\d+', '', text)

# Remove newline characters
  text = re.sub(r'\n', ' ', text)

  return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train,y_test = train_test_split(x ,y , test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
lr_pred = LR.predict(xv_test)
```

```python
LR.score(xv_test , y_test )
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test , lr_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train , y_train)
```

```python
dc_pred = DTC.predict(xv_test)
```

```python
DTC.score(xv_test , y_test)
```

```python
print(classification_report(y_test , dc_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train , y_train)
```

```python
rfc_pred = rfc.predict(xv_test)
```

```python
rfc.score(xv_test , y_test)
```

```python
print(classification_report(y_test , rfc_pred))
```

```python
# Function to clean and predict news
def predict_news(news_text):
    # Clean input text using the same preprocessing function
    cleaned_text = wordopt(news_text)

    # Convert the cleaned text into TF-IDF vector
    vectorized_text = vectorization.transform([cleaned_text])

    # Use your trained Logistic Regression model (you can change model if you want)
    prediction = LR.predict(vectorized_text)

    # Display the result
    if prediction == 0:
        print("\033[91m📰 The news is FAKE!\033[0m")
    else:
        print("\033[92m✅ The news is REAL!\033[0m")


# -------- User Interactive Input --------
while True:
    user_input = input("\nEnter a news article or headline (or type 'exit' to stop):\n> ")
    if user_input.lower() == 'exit':
        print("Exiting prediction tool... 👋")
        break
    predict_news(user_input)
```
